```python

data = [1121, "Jackie Grainger", 22.22,
1122, "Jignesh Thrakkar", 25.25,
1127, "Dion Green", 28.75, False,
24.32, 1132, "Jacob Gerber",
"Sarah Sanderson", 23.45, 1137, True,
"Brandon Heck", 1138, 25.84, True,
1152, "David Toma", 22.65,
23.75, 1157, "Charles King", False,
"Jackie Grainger", 1121, 22.22, False,
22.65, 1152, "David Toma"]

# declare 3 empty lists to store employee_names, employee_numbers, employee_hourly_wage details respectively
employee_names, employee_numbers, employee_hourly_wage = [], [], []

# iterate through each element in the data list
for i in data:
    # if the elment is a boolean value, continue the iterations by skipping the value
    if i == True or i == False:
        continue
        
    # if the elment is a string
    elif isinstance(i, str):
        # if the value does not exist already in the list
        if i not in employee_names:
            # append the value to employee_names list
            employee_names.append(i)
            
    # if the elment is a number      
    elif isinstance(i, int):
        # if the value does not exist already in the list
        if i not in employee_names: 
            # append the value to employee_numbers list
            employee_numbers.append(i)
            
    # if the elment is a decimal or float value 
    elif isinstance(i, float):
        # if the value does not exist already in the list
        if i not in employee_hourly_wage:
            # append the value to employee_hourly_wage list
            employee_hourly_wage.append(i)
        
print("Emplyee names: ",employee_names)
print("Emplyee numbers: ",employee_numbers)
print("Emplyee Hourly wages: ", employee_hourly_wage)

# declare a empty list to store the wages after providing 30% benefits
total_hourly_rate = []

# iterate through each wage in the employee_hourly_wage list
for wage in employee_hourly_wage:
    # multiply the current value by 1.3 and store it in total_hourly_rate list
    total_hourly_rate.append(wage * 1.3)
    
print("Total hourly rate wages: ", total_hourly_rate)

# find the maximum vaue in the total_hourly_rate and store it in a variable named 'highest_total_hourly_rate'
highest_total_hourly_rate = max(total_hourly_rate) 

# if the highest_total_hourly_rate value is more than 37.30, print an warning message
if highest_total_hourly_rate > 37.30:
    print("Warning!, Someone's salary may be a budget concern")
    
# declare a empty list to store the under paid wages
underpaid_salaries = []

# iterate through each wage in the total_hourly_rate list
for wage in total_hourly_rate:
    # if the value is between 28.15 and 30.65
    if wage > 28.15 and wage < 30.65 :
        # store the wage value in underpaid_salaries list
        underpaid_salaries.append(wage)
        
print("Underpaid salaries: ", underpaid_salaries)

# declare a empty list to store the raise wages of the under paid wages
company_raises = []

# iterate through each wage in the total_hourly_rate list
for wage in total_hourly_rate:
    if wage > 22 and wage < 24:         # if the hourly rate is between 22 and 24 
        raise_amount = wage * 0.05      # wage will be raised by 5%, so multiply the value by 0.05
    elif wage > 24 and wage < 26:       # if the hourly rate is between 24 and 26 
        raise_amount = wage * 0.04      # wage will be raised by 4%, so multiply the value by 0.04
    elif wage > 26 and wage < 28:       # if the hourly rate is between 26 and 28
        raise_amount = wage * 0.03      # wage will be raised by 3%, so multiply the value by 0.03
    else:                               # in all other cases
        raise_amount = wage * 0.02      # wage will be raised by 2%, so multiply the value by 0.02
    
    # store the total of wage and raise_ amount to the company_rise list
    company_raises.append(wage + raise_amount)
    
print("Company raised salaries: ", company_raises)
```

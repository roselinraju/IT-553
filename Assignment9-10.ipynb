```python
import re
# base class
class Individual:
    def __init__(self,name,emailAddress):
        self.name = name
        self.emailAddress = emailAddress

    def __str__(self):
        return "Name: " +self.name +",Email: "+self.emailAddress

# child  class Student inherits from Individual
class Student(Individual):
    def __init__(self,name,emailAddress,studentID,studyProgram):
        # call super class constructor
        super(Student, self).__init__(name,emailAddress)
        self.studentID = studentID
        self.studyProgram = studyProgram

    def __str__(self):
        return "Student ID: " +str(self.studentID) +",Program: "+self.studyProgram + ","+super(Student, self).__str__()

# child  class Instructor inherits from Individual
class Instructor(Individual):
    def __init__(self,name,emailAddress,instructorID,lastInstitution,highestDegree):
        # call super class constructor
        super(Instructor, self).__init__(name,emailAddress)
        self.instructorID = instructorID
        self.lastInstitution = lastInstitution
        self.highestDegree = highestDegree

    def __str__(self):
        return "Instructor ID: " +str(self.instructorID) +",Last Graduated institution: "+self.lastInstitution+", Highest Degree: "+self.highestDegree + ","+super(Instructor, self).__str__()

# class called "Validator" that has methods attached to it that validate submitted information.
class Validator:
    @staticmethod
    def valid_name(name):

        patternsContain = '^[a-zA-Z]*$'
        if re.search(patternsContain,name):
            return True
        else:
            return  False

    @staticmethod
    def valid_email(email):
        patterns = '^[a-zA-Z0-9@.]*$'
        if re.search(patterns, email):
            return True
        else:
            return False

    @staticmethod
    def valid_studentID(studentID):
        try:
            if ((studentID != "") and len(str(studentID)) <= 7 and (isinstance(int(studentID), int))):
                return True
        except:
            print("StudentId should be number")


    @staticmethod
    def valid_instructorID(instructorID):
        try:
            if ((instructorID != "") and len(str(instructorID)) <= 5 and (isinstance(int(instructorID), int))):
                return True
        except:
            print("Instructor ID should be number")

# display the details of the list
def displayInformation(collegeRecords):
    for index in range(len(collegeRecords)):
        print(collegeRecords[index].__str__())


if __name__ == '__main__':
    college_records = []
    type = ''
    # continue until the user enters done
    while type != 'done':
        # get type of individal
        type = input("Enter 'S' for Student or 'I' for Instructor or 'done' to stop : ")
        if type.lower() == 's' or type.lower() == 'i':
            while (1):
                name = input("Enter name of the individual: ")
                if Validator.valid_name(name):
                    break
                else:
                    print("Enter valid individual name")
            while (1):
                email = input("Enter email of the individual: ")
                if Validator.valid_email(email) == True:
                    break
                else:
                    print("Enter valid individual email")
            if type.lower() == 's':
                while (1):
                    studentId = input("Enter student ID: ")
                    if Validator.valid_studentID(studentId) == True:
                        break
                    else:
                        print("Enter valid student ID")
                program = input("Enter program of study: ")
                student = Student(name,email,studentId,program)
                college_records.append(student)
            elif type.lower() == 'i':
                while (1):
                    insId = input("Enter instructor ID: ")
                    if Validator.valid_instructorID(insId) == True:
                        break
                    else:
                        print("Enter valid instructor id")
                lastIns = input("Enter last institution they graduated: ")
                highDeg = input("Enter highest degree earned: ")
                instructor = Instructor(name,email,insId,lastIns,highDeg)
                college_records.append(instructor)
        elif type.lower() == 'done':
            displayInformation(college_records)
        else:
            print("Enter valid individual")
```
